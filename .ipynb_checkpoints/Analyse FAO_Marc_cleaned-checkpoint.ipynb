### Calcul du nombre de kcal nécessaires
L’apport nutritionnel recommandé (ANR) varie d'un individu à l'autre (dépendant de son âge, de son activité, de son sexe) et d'un pays à l'autre.
J'aurais pu prendre les recommandations générales (~2200kcal/jour/pers), mais comme il y a des différences entre homme et femme, j'ai voulu faire un peu d'exercice python :) Je n'ai pas tenu compte de l'âge moyen. Le résultat trouvé se rapproche des 2 329 kcal déterminées par la FAO comme référence pour leur étude présentée à la page 82 de ce [document](https://www.fao.org/3/ca9692fr/ca9692fr.pdf)<br>

Je me suis basée sur les indications suivantes :<br>
"Pour les hommes, le besoin énergétique moyen se situe entre 2400 et 2700 kcal. Pour les femmes, la recommandation est de 2000 à 2200 kcal." [source](https://www.la-vie-naturelle.com/blog/post/combien-de-calories-par-jour)<br>
"Sur 1000 personnes, 504 sont des hommes (50,4 %) et 496, des femmes (49,6 %)" [source](https://information.tv5monde.com/terriennes/78-milliards-d-etres-humains-mais-combien-de-femmes-2547)

Méthode :
1. Construction d'un dataframe dédié avec les données récoltées
2. Calcul de la moyenne homme, et de la moyenne femme
3. Calcul de la moyenne homme puis femme en appliquant le pourcentage de répartition homme/femme aux valeurs précédemment trouvées
4. Addition des 2 valeurs, ce qui nous donnera les besoins moyens pour un humain 

```python
df_repartition=pd.DataFrame({'kcal_bas' : [2400, 2000],'kcal_haut' : [2700, 2200],'percent': [50.4,49.6]},
                            index=["homme","femme"])
df_repartition
```

```python
mean_kcal_need_m=df_repartition.loc["homme",['kcal_bas','kcal_haut']].mean()
print(f'La moyenne d\'apport journalier recommandé pour un homme est de {mean_kcal_need_m:2} kcal')
```

```python
mean_kcal_need_w=df_repartition.loc["femme",['kcal_bas','kcal_haut']].mean()
print(f"La moyenne d'apport journalier recommandé pour une femme est de {mean_kcal_need_w:2} kcal")
```

```python
distrib_mean_kcal_m=mean_kcal_need_m*(df_repartition.loc['homme','percent']/100) # = 1285.2
distrib_mean_kcal_w=mean_kcal_need_w*(df_repartition.loc['femme','percent']/100) # = 1041.6
mean_kcal_need_human=distrib_mean_kcal_m+distrib_mean_kcal_w
print(f"La moyenne d'apport journalier recommandé pour un humain est de {mean_kcal_need_human} kcal")
```

Méthode :<br>
- Utilisation d'une fonction de comparaison
- En cas d'écart, recherche de potentiels pays qui pourraient correspondre.
- Correction


```python
# Création d'une liste des pays uniques (issue du fichier population) et de son dataframe
pop_unique_country_list=df_pop.Zone.unique()
df_pop_unique_country=pd.DataFrame(pop_unique_country_list, columns=["Zone"])
print(len(df_pop_unique_country), len(pop_unique_country_list))
```

```python
'''
Fonction compare_content_dataframe 

Paramètres :
    - un dataframe, celui dans lequel se trouvent les données à comparer.
    - une liste qui contient les valeurs de l'autre fichier à comparer avec le dataframe.

Actions :
    On compare une liste à un dataframe contenant une colonne.
    - S'il y a correspondance, on stock "Oui" dans la liste.
    - Sinon, on stocke "Non"

    A la fin de la boucle, on a une liste de "oui" et/ou "non".
    On va la coller dans une nouvelle colonne dans le dataframe contenant la liste des pays.
    - Oui = on a trouvé une correspondane entre la liste et le dataframe. 
    - Non = l'inverse

Note : on stocke les oui et les non pour conserver la taille du fichier.

Retour:
    Le dataframe contenant 2 colonnes : le nom du pays, et son statut ("Oui"/"Non").
'''
def compare_content_dataframe(dataframe, list_to_compare):
    array = []
    new_df=dataframe.copy()
    for i in range  (0, len(dataframe)):
        for pays in dataframe.values[i]:
            if pays in list_to_compare:
                array.append("Oui")
            else:
                array.append("Non")  
    new_df["Present dans la liste"]=array
    return new_df
```

```python
'''
Fonction search_for_potential_matching

Paramètres : 
    - Les 2 dataframes et les 2 listes de données uniques

Actions : 
    - Pour chaque entrée dans le premier dataframe, on regarde si on retrouve le nom dans la seconde liste.
        - Si on trouve une correspondance dans la liste, on stock le résultat de cette liste dans un tableau.
    - Pour chaque entrée dans le second dataframe, on regarde si on retrouve le nom dans la première liste.
        - Si on trouve une correspondance dans la liste, on stock le résultat du second dataframe dans un tableau.
Retour : 
     - Une liste de résultats contenant le nom des informations manquantes
'''

def search_for_potential_matching (missing_data_from_df1, missing_data_from_df2, list2, list1):
    potential_match=[]
    for missing_data in missing_data_from_df1:
        for element in list2:
            if missing_data.lower() in element.lower():
                print(f"Potential matching in second file for {element} :",element)
                potential_match.append(element)
    
    for missing_data2 in missing_data_from_df2:    
        for element2 in list1:
            if missing_data2.lower() in element2.lower():
                print(f"Potential matching in first file for {element2}:",missing_data2)
                potential_match.append(missing_data2)    
    return potential_match   
```

Comparaison des fichiers<br>
On commence par comparer le fichier le plus court. <br>
S'il n'y a pas de résultat, c'est qu'on a trouvé toutes les correspondances.<br>
Sinon, on compare le fichier le plus long avec le plus court afin de récupérer la liste des pays qui n'ont pas trouvé de correspondance.<br>
Cette liste va nous servir pour chercher les potentiels noms identiques.



```python
df_missing_countries_from_avail=compare_content_dataframe(df_avail_unique_country,pop_unique_country_list)

if len(df_missing_countries_from_avail[df_missing_countries_from_avail["Present dans la liste"]=="Non"])!=0:
    df_missing_countries_from_pop=compare_content_dataframe(df_pop_unique_country,avail_unique_country_list)
    print(f""" Les pays non trouvés sont les suivants :
    {','.join(list(df_missing_countries_from_avail[df_missing_countries_from_avail["Present dans la liste"]=="Non"]["Zone"]))}""")
    potential_countries = search_for_potential_matching(df_missing_countries_from_avail[df_missing_countries_from_avail["Present dans la liste"]=="Non"]["Zone"]\
                    , df_missing_countries_from_pop[df_missing_countries_from_pop["Present dans la liste"]=="Non"]["Zone"],\
                    pop_unique_country_list,avail_unique_country_list)
else:
    print("Tous les pays présents dans dispo_alimentaire sont présents dans le fichier population.")
```

### Calcul du nombre de personnes qu'on peut nourrir à partir de la production journalière 
- Création de 2 dataframes ne contenant que les informations qui nous intéressent
    - dispo_alimentaire : *Zone*, *Origine*,*Disponibilité de protéines en quantité (g/personne/jour)* et *Disponibilité alimentaire (Kcal/personne/jour)* en regroupant les informations par pays et par origine
    - population : filtre sur l'année 2017

- Merge de ces 2 dataframes avec uniquement les valeurs non vides (pas de jointure)
- Factorisation de la méthode de calcul pour pouvoir l'utiliser pour les différents types d'*Origine* et d'énergie (calorie, protéine)


Regroupement par pays et origine

```python
df_avail_energy=df_avail.groupby(['Zone','Origine'], as_index=False).agg({'Disponibilité alimentaire (Kcal/personne/jour)':'sum',"Disponibilité de protéines en quantité (g/personne/jour)":'sum'})
df_avail_energy.head(5)
```

```python
df_pop_2017=df_pop.loc[(df_pop["Année"]==2017),["Zone","Valeur"]].copy()
df_pop_2017.head()
```

```python
df_prod_by_pop=pd.merge(df_pop_2017, df_avail_energy, left_on= "Zone", right_on= "Zone")
df_prod_by_pop.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)': 'Dispo alim (kcal/pers/jour)', "Zone":"Pays","Valeur":"Population" ,"Disponibilité de protéines en quantité (g/personne/jour)":"Dispo prot (g/pers/jour)"}, inplace=True)
df_prod_by_pop["Ratio production calorique"]=df_prod_by_pop["Dispo alim (kcal/pers/jour)"]/mean_kcal_need_human
df_prod_by_pop["Total prod kcal/jour/pers"]=df_prod_by_pop["Population"]*df_prod_by_pop["Ratio production calorique"]
df_prod_by_pop["Ratio production protéique"]=df_prod_by_pop["Dispo prot (g/pers/jour)"]/prot_need
df_prod_by_pop["Total prod prot(g)/jour/pers"]=df_prod_by_pop["Population"]*df_prod_by_pop["Ratio production protéique"]
df_prod_by_pop.head()
```

```python
'''
Fonction people_who_could_be_fed_by_energy 
La fonction prend en variable 
- un dataframe
- le type d'"énergie" : calories ou protéines
- l'origine des denrées, animale ou vegetale, qui est facultatif.
    - Si la valeur est vide, on ne fait pas de filtre sur la colonne "origine"
    
La fonction retourne une phrase contenant 
- Le nombre de personnes qu'on peut nourrir à partir des informations passées en paramètre
- Le pourcentage de la population concerné
'''

def people_who_could_be_fed(dataframe,energy,origine=""):
    global df_origine_prod
    global world_energy_avail
    
    if origine in["vegetale","animale"] :
        df_origine_prod=dataframe[dataframe["Origine"]==origine].copy()
    elif origine=="":
        df_origine_prod=dataframe
        origine="vegetale et animale"
        
    if energy=="protéines":
        world_energy_avail=df_origine_prod["Total prod prot(g)/jour/pers"].sum()
    elif energy=="calories":
        world_energy_avail=df_origine_prod["Total prod kcal/jour/pers"].sum()
    percent_world_energy_avail=world_energy_avail/population_2017 

    
    return f"Nous avons suffisamment de {energy} d'origine {origine} pour nourrir {world_energy_avail/1000000000:.2} milliards de personnes, soit {percent_world_energy_avail:.2%} de la population."
    
```

## Utilisation des céréales

Etapes :

5. Les céréales renommées :
    - "Mil" devient "Millet" (*"Ce terme est un synonyme, employé surtout en Afrique, de millet ou de sorgho.)* : comme on a Sorgho, on extrapole et transforme Mil en Millet

    - "Mélanges de céréales" devient "Miscellanees"
    - Le quinoa et le sarrasin sont considérées comme des pseudocéréales, on les retrouve probablement sous "Autres céréales" dans la disponibilité alimentaire.

Construction du dataframe contenant uniquement les information de dispo_alimentaire sur nos céréales

```python
df_cereals_avail=pd.DataFrame()
for i in range (0,len(df_cereal_list_compare)):
    for cereal in list(df_cereal_list_compare.values[i]):
        #df_avail_cereals=df_avail_cereals.append(df_avail[(df_avail["Produit"])==element])
        df_cereals_avail=pd.concat([df_cereals_avail,df_avail[(df_avail["Produit"])==cereal]]) 
df_cereals_avail.head()
```

Vérification qu'il n'y a bien que des lignes avec nos céréales.

```python
df_cereals_avail.Produit.unique()
```

On peut maintenant calculer la part de céréales pour l'alimentation humaine et animale

```python
animal_food = df_cereals_avail["Aliments pour animaux"].sum()
human_food=df_cereals_avail["Nourriture"].sum()
total_food = animal_food+human_food
prop_animal=animal_food/total_food
prop_human=human_food/total_food
print(f"La production mondiale de céréales (dédiée uniquement à l'alimentation) sert à nourrir {prop_animal:.2%} des animaux et {prop_human:.2%} des humains.")
```

### Conclusion
Un peu moins de la moitié de la production mondiale de céréales est dédiée à l'alimentation des animaux.
En diminuant la production de nourriture d'origine animale, on libérerait instantanément une partie de la production de céréales qu'on pourrait consacrer à la nourriture pour les humains.

## Analyse de l'utilisation des produits en Thaïlande (dont manioc) et de leur niveau d'exportation 

Etapes :
- Calcul de la population thaïlandaise
- Calcul de la part de thaïlandais en sous-nutrition
- Création d'un tableau avec tous les aliments produits en Thaïlande auquel on ajoute 2 colonnes contenant 
    - le ratio d'export (Exportation/Production) 
    - le ratio du produit dédié à la nourriture humaine (Nourriture / Dispo intérieure)

Analyse du tableau
- On filtre le tableau pour ne garder que  
    - un taux arbitraire > 70% sur le ratio d'export
    - un faible ratio d'alimentation pour l'Homme, < 50%

```python
thai_pop=df_pop_2017[df_pop_2017.Zone=="Thaïlande"]["Valeur"].values[0]
print(f"Il y a {thai_pop/1000000:.4} millions de thaïlandais.")
```

```python
thai_pop_undernut=df_undernut[(df_undernut["Zone"]=="Thaïlande")&(df_undernut["Année"]==2017)].Valeur.values[0]
percent_thai_pop_undernut=thai_pop_undernut/thai_pop
print(f"{thai_pop_undernut/1000000} millions de thaïlandais sont en sous-nutrition ({percent_thai_pop_undernut:.2%})")
```

```python
df_avail_thailande_global=df_avail.loc[(df_avail['Zone'] == 'Thaïlande')].copy()
df_avail_thailande_global.head()
```

Je supprime les colonnes inutiles et je créée celles dont j'ai besoin

```python
df_avail_thailande_global.drop(columns=['Disponibilité alimentaire en quantité (kg/personne/an)', 'Disponibilité de matière grasse en quantité (g/personne/jour)',
                                        "Disponibilité de protéines en quantité (g/personne/jour)","Disponibilité alimentaire (Kcal/personne/jour)"], inplace=True)
```

Pour éviter les erreurs dans les calculs (division par 0), je remplace les valeurs nulles par zéro dans les colonnes que je vais exploiter

```python
df_avail_thailande_global["Disponibilité intérieure"]=df_avail_thailande_global["Disponibilité intérieure"].fillna(0)
df_avail_thailande_global["Exportations - Quantité"]=df_avail_thailande_global["Exportations - Quantité"].fillna(0)
df_avail_thailande_global["Production"]=df_avail_thailande_global["Production"].fillna(0)
df_avail_thailande_global["Nourriture"]=df_avail_thailande_global["Nourriture"].fillna(0)
df_avail_thailande_global.head()
```

On calcule les ratio pour chaque ligne

```python
df_avail_thailande_global["Pourentage export"] = 0
df_avail_thailande_global["Pourcentage nourriture pour humain"]=0
for column, row in df_avail_thailande_global.iterrows():
    if (row["Production"]!=0):
        df_avail_thailande_global.loc[column,'Pourentage export'] = round(row['Exportations - Quantité']/row['Production'] * 100,2)
    if (row["Disponibilité intérieure"]!=0):
        df_avail_thailande_global.loc[column,'Pourcentage nourriture pour humain'] = round(row['Nourriture']/row['Disponibilité intérieure'] * 100,2)
        df_avail_thailande_global.loc[column,'Pourcentage nourriture pour animaux'] = round(row['Aliments pour animaux']/row['Disponibilité intérieure'] * 100,2)
df_avail_thailande_global.head()
```

On supprime les lignes ayant un pourcentage d'export inférieur ou égal à 70% et un pourcentage de nourriture dédiée aux humains de plus de 50%

```python
indexNames = df_avail_thailande_global[(df_avail_thailande_global["Pourentage export"] <=70) | (df_avail_thailande_global["Pourcentage nourriture pour humain"]>50)].index
df_avail_thailande_global.drop(indexNames, inplace=True)
```

```python
df_avail_thailande_global
```

```python
df_avail_manioc_thailande=df_avail_thailande_global[df_avail_thailande_global['Produit'] == 'Manioc']
```

```python
manioc_exp=df_avail_manioc_thailande["Pourentage export"].values[0]
manioc_for_human=df_avail_manioc_thailande["Pourcentage nourriture pour humain"].values[0]
print(f"""La Thaïlande exporte {manioc_exp}% de sa production de manioc.
Seulement {manioc_for_human}% de ce qu'il reste pour le pays est destiné à l'alimentation de l'Homme.""")
```

### Conclusion
La Thaïlande étant un pays en sous-nutrition devrait limiter ses exports de **manioc** afin de la destiner à sa population. D'autant plus que la part restante est très peu utilisée pour l'Homme.<br>
Le **manioc** est le seul aliment consommable par les humains qui massivement exporté. <br>
A noter toutefois que cette tubercule n'est pas un aliment sans reproche : il est très calorique et apporte peu de protéines, qui plus est de mauvaise qualité.<br>

Le **blé** est majoritairement utilisé pour les animaux mais n'est pas produit localement puisque la majorité de la disponibilité intérieure provient de l'importation.<br>

**L'huile** est majoritairement exportée mais est utilisée à d'autres fins que l'alimentation, et les **autres céréales** sont certainement cultivées pour les animaux.

## Exportation des pays en insécurité alimentaire
Essayons de voir le comportement des pays en insécurité alimentaire vis-à-vis de leurs exportations.

Création d'un dataframe sous-nutrition de l'année 2017 avec les valeurs nulles ou = à 0 supprimées

```python
countries_undernourished_2017_df=df_undernut[df_undernut.Année==2017].dropna()
countries_undernourished_2017_df.drop(countries_undernourished_2017_df[countries_undernourished_2017_df.Valeur==0].index, inplace=True)
countries_undernourished_2017_df.drop(columns="Année", inplace=True)
countries_undernourished_2017_df.head()
```

```python
df_avail.sort_values(by="Exportations - Quantité", ascending=False)
```

Suppression des na et 0 sur la colonne export

```python
df_avail_cleaned=df_avail.copy().dropna(subset=['Exportations - Quantité'])
df_avail_cleaned.drop(df_avail_cleaned[df_avail_cleaned["Exportations - Quantité"]==0].index, inplace=True)
df_avail_cleaned.sort_values(by="Exportations - Quantité", ascending=False)
```

Merge des 2 tableau pour se débarasser des pays ne figurant pas dans le dataframe dispo_alimentaire

```python
df_avail_cleaned=df_avail_cleaned.merge(countries_undernourished_2017_df).sort_values(by="Exportations - Quantité", ascending=False)
df_avail_cleaned.head()
```

Je regarde les 10 produits les plus exportés après avoir fait la somme des exportations pour chaque produit.

```python
most_exported_product_in_undernut_df=df_avail_cleaned.groupby("Produit", as_index=False).agg({"Exportations - Quantité" : 'sum', "Disponibilité intérieure":"sum"}).sort_values("Exportations - Quantité", ascending=False)
most_exported_product_in_undernut_df["Ratio export"]=round(most_exported_product_in_undernut_df["Exportations - Quantité"]/most_exported_product_in_undernut_df["Disponibilité intérieure"],2)
most_exported_product_in_undernut_top10_df=most_exported_product_in_undernut_df.head(10)
most_exported_product_in_undernut_top10_df
```

```python
fig = plt.figure()
ax = fig.add_axes([1, 0, 2, 1])
ax.bar(most_exported_product_in_undernut_top10_df["Produit"], most_exported_product_in_undernut_top10_df["Exportations - Quantité"])
#ax.bar(etiquettes, valeurs,
plt.xlabel('Produits')
plt.ylabel('Exportation')
plt.title('Top 10 des produits les plus exportés par les pays en insécurité alimentaire')
```

Je créé une liste à partir du résultat du dataframe, contenant uniquement les 10 produits, puis je filtre sur le dataframe nettoyé pour ne garder que les lignes dont le produit figure dans la liste.

```python
product_mainly_exported_list=list(most_exported_product_in_undernut_top10_df.Produit)
most_exported_product_df=df_avail_cleaned[df_avail_cleaned.Produit.isin(product_mainly_exported_list)].copy()
most_exported_product_df.head()
```

Je me fait un dataframe contenant les informations que je juge utile pour regarder ce que les pays font de ces produits pour l'alimentation

```python
most_exported_product_df.drop(columns=['Zone', 'Origine', 'Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)',\
                                        'Disponibilité de matière grasse en quantité (g/personne/jour)','Disponibilité de protéines en quantité (g/personne/jour)',\
                                        'Variation de stock', 'Valeur','Semences','Traitement'], inplace=True)
```

```python
most_exported_product_df=most_exported_product_df.groupby(by="Produit", as_index=False).agg({"Nourriture" : 'sum', "Aliments pour animaux" : 'sum', "Autres Utilisations" : 'sum',"Pertes" : 'sum',\
                                                                                             "Disponibilité intérieure" : 'sum'})

most_exported_product_df["Ratio nourriture humains/dispo int"]=round(most_exported_product_df["Nourriture"]/most_exported_product_df["Disponibilité intérieure"],2)
most_exported_product_df["Ratio nourriture animaux/dispo int"]=round(most_exported_product_df["Aliments pour animaux"]/most_exported_product_df["Disponibilité intérieure"],2)
most_exported_product_df["Utilisation animale>humaine?"]=most_exported_product_df["Ratio nourriture humains/dispo int"]<most_exported_product_df["Ratio nourriture animaux/dispo int"]
most_exported_product_df.sort_values(["Ratio nourriture animaux/dispo int"], ascending=False)
```

```python
# Position sur l'axe des x pour chaque étiquette
position = np.arange(len(most_exported_product_df["Produit"].values))
largeur = .3
# Création de la figure et d'un set de sous-graphiques
fig = plt.figure()
ax = fig.add_axes([1, 0, 2.3, 1.2])

r1 = ax.bar(position - largeur/2, most_exported_product_df["Ratio nourriture humains/dispo int"], largeur, color='#d420d7')
r2 = ax.bar(position + largeur/2, most_exported_product_df["Ratio nourriture animaux/dispo int"], largeur,color='#37ecec')

#Noms des axes
plt.xlabel('Produits')
plt.ylabel('Ratio')

# Modification des marques sur l'axe des x et de leurs étiquettes => on remplace par les noms de produits
ax.set_xticks(position,most_exported_product_df["Produit"])

plt.title('Répartition nourriture animaux/humains par rapport à la disponibilité intérieure')

plt.legend(['Humains', 'Animaux'])
plt.show()
```

### Conclusion
**Les produits exportés**<br>
Les produits les plus exportés par les pays en insécurité alimentaire sont des aliments qui pourraient servir pour nourrir leurs habitants dans le besoin.

Le soja par exemple est un excellent aliment, apportant beaucoup de protéines et étant peu calorique. Mais il doit être transformé pour être propre à la consommation.

Le riz, "constitue la base de l’alimentation pour environ la moitié de la population du globe. On rapporte même qu’environ 23% de toutes les calories consommées dans le monde proviendraient du riz."

Le manioc, massivement exporté, ne constitue pas un aliment équilibré : il est très calorique mais apporte peu de protéine. C'est une solution contre la faim mais pas de l'équilibre alimentaire, qui est tout aussi important.
De surcroît, le peu de protéines disponibles est de mauvaise qualité, et provoque des malnutritions protéiques.
"On constate d'ailleurs que les cartes de malnutrition protéique (kwashiorkor) coïncident avec les régions où les enfants sont principalement nourris à base de manioc et de bananes."

[Source sur les valeurs nutritives du manioc](https://devsante.org/articles/le-manioc)<br>
[Source qualités nutritives des aliments](https://www.passeportsante.net/portail/aliments)

**L'alimentation pour les animaux**<br>
Le maïs, qui est l'aliment le plus exporté par les pays en état de sous-alimentation est utilisé en majorité pour l'alimentation animale, à presque 50%, contre 35 pour l'Homme.<br>
En diminuant l'élevage et en consacrant une partie de ce qui est dédié à la nourriture animale pour l'homme, il y a un levier d'amélioration pour l'insécurité alimentaire.


**Autres**<br>
Le soja n'est utilisé ni pour l'alimentation des humains ni pour celle des animaux. La consommation de soja se fait via des produits transformés, la graine de soja pure n'est pas exploitée.
